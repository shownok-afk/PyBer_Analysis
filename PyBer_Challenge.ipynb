### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
%matplotlib notebook
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import os


# File to Load
city_data_to_load = os.path.join("Resources", "city_data.csv")
ride_data_to_load = os.path.join("Resources", "ride_data.csv")

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
ride_count
```

```python
# 2. Get the total drivers for each city type
driver_count = city_data_df.groupby(["type"]).sum()["driver_count"]
driver_count
```

```python
#  3. Get the total amount of fares for each city type
fare_count = pyber_data_df.groupby(["type"]).sum()["fare"]
fare_count
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = fare_count / ride_count
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = fare_count / driver_count
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    'Total Rides':ride_count,
    'Total Drivers':driver_count,
    'Total Fares':fare_count,
    'Average Fare per Ride':average_fare_per_ride,
    'Average Fare per Driver':average_fare_per_driver
})

# pyber_summary_df.set_index("type", inplace=True)

#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None

#  8. Format the columns.
# Format the "Total Rides" to have the comma for a thousands separator.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)

# Format the "Total Drivers" to have the comma for a thousands separator.
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)

# Format the "Total Fares" to have the comma for a thousands separator, a decimal separator and a "$".
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)

# Format the "Average Fare per Ride" to have the comma for a thousands separator, a decimal separator and a "$".
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)

# Format the "Average Fare per Driver" to have the comma for a thousands separator, a decimal separator and a "$".
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)

pyber_summary_df

```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fare_based_on_type_date_df = pyber_data_df.groupby(['type','date']).sum()['fare']
fare_based_on_type_date_df.head()
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fare_based_on_type_date_df = fare_based_on_type_date_df.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date.
fare_based_on_type_date_pivot_df = fare_based_on_type_date_df.pivot(index = 'date', columns = 'type', values = 'fare')
fare_based_on_type_date_pivot_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fare_based_on_type_new_date_range_df = fare_based_on_type_date_pivot_df.loc['2019-01-01':'2019-04-29']
fare_based_on_type_new_date_range_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
fare_based_on_type_new_date_range_df.index = pd.to_datetime(fare_based_on_type_new_date_range_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
fare_based_on_type_new_date_range_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fare_based_on_type_new_week_df = fare_based_on_type_new_date_range_df.resample("w").sum()
fare_based_on_type_new_week_df
```

```python
fare_based_on_type_new_month_df = fare_based_on_type_new_week_df
fare_based_on_type_new_month_df = pd.DataFrame({'date': pd.date_range('2019-01-01', '2020-01-01', freq='1D')})
fare_based_on_type_new_month_df['Month'] = fare_based_on_type_new_month_df.date.dt.month_name()
fare_based_on_type_new_month_df.head()
```

```python
# Combine the data into a single dataset
fare_based_on_type_new_date_month_df = pd.merge(fare_based_on_type_new_week_df, fare_based_on_type_new_month_df, how="left", on=["date", "date"])
fare_based_on_type_new_date_month_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
ax = fare_based_on_type_new_week_df.plot(figsize = (20, 6))
ax.set_title("Total Fare by City Type")
ax.set_ylabel("Fare ($USD)")
ax.legend(loc = 10)

# Save Figure
plt.savefig("analysis/PyBer_fare_summary.png")

plt.show()
```
